### Write a function that can return the track record of 2 teams against each other (Ipl)

```python
def IplTracker(team1, team2, ipl):
    # Filter the DataFrame for matches involving the two teams
    team1_vs_team2 = ipl[   (  (ipl['Team1'] == team1) & (ipl['Team2'] == team2)  ) | 
                         ( (ipl['Team1'] == team2) & (ipl['Team2'] == team1) )         ]

    # Count the number of matches played
    total_matches = team1_vs_team2.shape[0]

    # Count the number of matches won by each team
    team1_wins = team1_vs_team2[team1_vs_team2['WinningTeam'] == team1].shape[0]
    team2_wins = team1_vs_team2[team1_vs_team2['WinningTeam'] == team2].shape[0]

    # Calculate win percentages
    team1_win_percentage = (team1_wins / total_matches) * 100
    team2_win_percentage = (team2_wins / total_matches) * 100

    print(f'Total Matches Played: {total_matches}')
    print(f'{team1} Wins: {team1_wins} ({team1_win_percentage:.2f}%)')
    print(f'{team2} Wins: {team2_wins} ({team2_win_percentage:.2f}%)')
```

```python
IplTracker('Royal Challengers Bangalore', 'Mumbai Indians', ipl)
```
