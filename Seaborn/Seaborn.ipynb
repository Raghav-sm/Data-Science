# Seaborn

It's a Data visualixation library built on matplotlib

Importing modules

```python
import pandas as pd
import numpy as np 
%matplotlib inline 
import matplotlib.pyplot as plt 
import seaborn as sns
%reload_ext autoreload
%autoreload 2
```

# Importing Data

```python
print(sns.get_dataset_names()) #built in data sets
```

```python
crash_df = sns.load_dataset('car_crashes')
crash_df
```

# Distribution plot

```python
sns.distplot(crash_df['not_distracted'] , kde = True,bins = 30);
```

# Joint plot

Basically comapres to distribution plot

```python
sns.jointplot( data = crash_df , x = 'ins_losses' , y = 'alcohol' );
# sns.jointplot?
# kde = kernel density estimation or we could have given kind = reg , hex
```

# KDE plot

```python
sns.kdeplot(crash_df['alcohol']);
```

# Pair plot

This Basically plots the relatonship between all the numerical data values

```python
sns.pairplot(crash_df);
```

```python
tips_df = sns.load_dataset('tips')
tips_df.head()
```

```python
sns.pairplot(tips_df , hue = 'sex' , palette = 'Blues');
```

# Rug plot

```python
sns.rugplot(tips_df['tip'] );
```

# styling

```python
heart_df = pd.read_csv(r"E:/conda_project/Seaborn/heart-disease.csv")
heart_df.head()
```

```python

sns.set_style('darkgrid') # style : dict, or one of {darkgrid, whitegrid, dark, white, ticks}
sns.jointplot(data = heart_df , x = 'age' , y = 'chol' , kind = 'reg');
sns.set_context('poster' , font_scale = 1)
```

# Categorical Plots 
## Bar plots


```python
sns.barplot( x ='sex' , y = 'chol' , data = heart_df , estimator = np.std);
#sns.barplot( x ='sex' , y = 'chol' , data = heart_df , estimator = np.var);
#sns.barplot( x ='sex' , y = 'chol' , data = heart_df , estimator = np.median);
```

## countplot

```python
sns.countplot( x = 'sex' , data = heart_df);
```

### Boxplot

```python
tips_df.head()
```

```python
sns.boxplot(data = tips_df , x = "day" , y = "total_bill" , hue = "sex");
```

## Palette

```python
plt.figure(figsize = (10,8));
sns.set_style('dark')
sns.set_context('talk')
sns.stripplot( x = 'day' , y = 'total_bill' , data = tips_df , hue = 'sex' , palette = 'magma')
plt.legend(loc = 1 )
```

## Heatmaps

```python
plt.figure(figsize = (10 ,8))
sns.set_style('dark')
sns.set_context('paper' , font_scale = 1.4 )

heart_matrix = heart_df.corr()
sns.heatmap( data = heart_matrix , annot =True , cmap = "Blues" , fmt = '.2f')
```
