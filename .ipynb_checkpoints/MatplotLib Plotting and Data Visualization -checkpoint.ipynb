# Matplotlib

Matplotlib is a Python library that lets experienced coders create detailed and customizable visualizations, from basic charts to complex data plots, for effective data analysis and presentation. It's a powerful tool to make data speak visually.

**Document format :** After each cell ,there will be an explanation of each line of code.  
At first let us import the module.


````python
%matplotlib inline 
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np
````

Certainly, here it is in bullet points with corrected markdown syntax:

- `%matplotlib inline`: This command tells the Jupyter Notebook to display Matplotlib plots directly in the output of the code cells.
- The next three lines import important Python libraries:
  1. `Matplotlib` for creating graphs.
  2. `Pandas` for data handling.
  3. `NumPy` for numerical operations, making these tools available for your code.

1. **Pyplot method**    `plt.plot();` is telling Matplotlib to create a plot, like drawing a blank canvas for a graph, but you'll need to add data and customize it further to see something meaningful.

````python
x = [ 1 ,2 ,3, 4]
y = [ 11 ,22 ,33 ,44]
plt.plot(x , y); #Stateless way of plotting , similar to matlab
````

Alternative way to do it :-
- `fig, ax = plt.subplots()`: This line creates a figure and an axis to draw your plot on.
- `ax.plot(x, y)`: It plots data from arrays `x` and `y` on the axis, allowing you to visualize your data effectively.

````python
fig , ax = plt.subplots()
ax.plot(x , [10,20,30,40]);
````

### Matplotlib workflow  
1. Importing matplotlib
```
%matplotlib inline  
import matplotlib.pylot as plt  
```
2. Preparing data using Numpy ,pandas or importing it directly
```python
x = [1,2,3,4]
y = [10,20,30,40]
```
3. Plotting it  
```python
ax , fig = plt.subplots(figsize=(10,10))
ax.plot(x , y )
```
4. Customizing plot  
```python
ax.set(title = "Matplotlib" ,
       xlabel = "X-Axis" , 
       ylabel = "Y-Axis")
```
5. Saving
```python
fig.savegig("wherever you want")
```

````python
x = np.linspace(0,10,100)
````

````python
fig ,ax = plt.subplots()
ax.plot(x , x**2);# Default is lineplot
````

````python
#Scatterplot
fig , ax = plt.subplots()
ax.scatter( x , np.exp(x));
````

````python
#using Dictionary to create plots
fruits = { "Apple " : 30,
           "Banana" : 5,
           "Orange" : 14 } 
fig , ax = plt.subplots()
ax.bar(list(fruits.keys()) , list(fruits.values())); 
ax.set(title = "Fruit Store" , 
       ylabel = 'Number of fruits',
       xlabel = 'Fruit')
````

````python

fig , ax = plt.subplots()
ax.barh(list(fruits.keys()) , list(fruits.values()));
````

````python
#histogram
x = np.random.randn(1000)
fig , ax = plt.subplots()
ax.hist( x );
````

````python
# Thr is two options for subplots
#option 1 
fig , ((ax1 ,ax2) , (ax3,ax4)) = plt.subplots( nrows = 2 ,
                                               ncols = 2 ,
                                             figsize=(10,5))
#plotting Individually 
ax1.plot( x , x /2);
ax2.scatter(np.random.random(10) , np.random.random(10));
ax3.bar(list(fruits.keys()) , list(fruits.values()));
ax4.hist(np.random.randint(0,10,2));
````

````python
# option 2 
fig , ax = plt.subplots(nrows = 2 ,
                        ncols = 2 ,
                        figsize= (10, 5))
ax[0,0 ].plot( x , x/2 );
ax[0,1].scatter(np.random.random(10) , np.random.random(10));
ax[1,0].bar(list(fruits.keys()) , list(fruits.values()));
ax[1,1].hist(np.random.randint(0,10,2));
````

````python
#plotting using pandas
car_sales = pd.read_csv('car-sales.csv')
car_sales
````

````python
car_sales = pd.read_csv('car-sales.csv')
car_sales
````

````python
#removing the dollar size
car_sales['Price'] = car_sales['Price'].str.replace('[\$\,\.]' ,'',regex = True)
car_sales
````

````python
#removing last two zeros
car_sales['Price'] = car_sales['Price'].str[: -2]
car_sales
````

````python
car_sales['Sale Date'] = pd.date_range(start = '23/10/23' , periods = len(car_sales))
car_sales
````

Now we have to add all the pice ,so we can use ```cumsum()```,but before that we need to create the object type to integer type 

````python
car_sales['Total Sale'] = car_sales['Price'].astype(int).cumsum()
car_sales
````

Now plotting the total sales

````python
car_sales.plot( x = 'Sale Date' , y = 'Total Sale');
````

````python
car_sales.plot( x = 'Odometer (KM)' , y = 'Total Sale' , kind = 'scatter');
````
